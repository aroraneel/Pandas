```python
# TASK 28: Create a duplicate row by adding an existing row again
df = pd.concat([df, df.iloc[[1]]])
df.tail()
```

```python
# TASK 29: Identify duplicate rows (True = duplicate)
df.duplicated()
```

```python
# TASK 30: Remove duplicate rows from the dataset
df.drop_duplicates(inplace=True)
df.shape
```

```python
# TASK 31: Use apply() to create a new column (Adult / Child based on Age)
def age_group(age):
    return "Adult" if age >= 18 else "Child"

df['AgeGroup'] = df['Age'].apply(age_group)
df[['Age', 'AgeGroup']]
```

```python
# TASK 32: Use map() to convert categorical Gender into numeric values
df['SexCode'] = df['owner_sex'].map({'male': 1, 'female': 0})
df[['owner_sex', 'SexCode']]
```

```python
# TASK 33: Use map() with function to classify income as High or Low
def fare_level(value):
    return "High" if value > df['personal_income'].median() else "Low"

df['FareLevel'] = df['personal_income'].map(fare_level)
df[['personal_income', 'FareLevel']]
```

```python
# TASK 34: Export the cleaned DataFrame to a CSV file
df.to_csv("titanic_cleaned.csv", index=False)
```
